#### Defining functions for .py file

```python
def dbfile_to_dataframe (dbfile):
    """
    Creates a SQL connection to the SQLite. This function reads SQL database files,
    executes SQL queries in python platform and extracts data for specific table and
    returns a pandas DataFrame
    
    Parameters
    ----------
    - dbfile: path of the database file
    
    
    Returns
    -------
    - modified DataFrame based on the user selected tables and columns
    """
    
    con = sqlite3.connect(dbfile)
    df=pd.read_sql_query("""SELECT * FROM source""", con)
    
    return df
```

```python
def merged_data(df1, df2):
    
    """This function combines two dataframes
    
    Parameters
    ----------
    - df1,df2: dataframes to be combined
    
    Returns
    -------
    - merged DataFrame
    """
    df_merged=pd.concat([df1, df2])
    
    return df_merged
```

### Visualization of data frequency in different organisms

```python
import matplotlib
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams.update({'font.size': 18})
```

```python
df_final=pd.read_csv('cattleplusrat_final.csv', sep=';')
```

```python
df_final.columns=['organism','chembl_id','canonical_smiles','t_half','f','auc','drug_uptake','ic50', 'cmax','vdss','cl','ld50','ppb','drug_metabolism','distribution','ed50','permeability','ratio','papp']
```

```python
df_final
```

```python
organisms=list(df_final['organism'].unique())
len(organisms)
```

```python
ld50=[]
ic50=[]
t_half=[]
papp=[]
f=[]
auc=[]
drug_uptake=[]
cmax=[]
vdss=[]
cl=[]
ppb=[]
ed50=[]


for i in organisms:
    organism=df_final[df_final['organism']==i]
    ld50.append(organism.ld50.count()/len(organism))
    ic50.append(organism.ic50.count()/len(organism))
    t_half.append(organism.t_half.count()/len(organism))
    papp.append(organism.papp.count()/len(organism))
    f.append(organism.f.count()/len(organism))
    auc.append(organism.auc.count()/len(organism))
    drug_uptake.append(organism.drug_uptake.count()/len(organism))
    cmax.append(organism.cmax.count()/len(organism))
    vdss.append(organism.vdss.count()/len(organism))
    cl.append(organism.cl.count()/len(organism))
    ppb.append(organism.ppb.count()/len(organism))
    ed50.append(organism.ed50.count()/len(organism))
    
```

```python
plt.figure(figsize=(25,15))
sns.barplot(x=ld50, y=organisms, color='yellow', alpha=0.5,label='LD50')
sns.barplot(x=ic50, y=organisms, color='green', alpha=0.5,label='IC50')
sns.barplot(x=t_half, y=organisms, color='blue', alpha=0.5,label='t1/2')
sns.barplot(x=papp, y=organisms, color='red', alpha=0.5,label='Papp')
sns.barplot(x=f, y=organisms, color='chartreuse', alpha=0.5,label='F')
sns.barplot(x=auc, y=organisms, color='maroon', alpha=0.5,label='AUC')
sns.barplot(x=drug_uptake, y=organisms, color='magenta', alpha=0.5,label='Drug_uptake')
sns.barplot(x=cmax, y=organisms, color='gray', alpha=0.5,label='Cmax')
sns.barplot(x=vdss, y=organisms, color='cyan', alpha=0.5,label='Vdss')
sns.barplot(x=cl, y=organisms, color='firebrick', alpha=0.5,label='CL')
sns.barplot(x=ppb, y=organisms, color='khaki', alpha=0.5,label='PPB')
sns.barplot(x=ed50, y=organisms, color='lightpink', alpha=0.5,label='ED50')

plt.legend(loc='best', fontsize=20)
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='medium')
matplotlib.rc('xtick', labelsize=25) 
matplotlib.rc('ytick', labelsize=25) 

plt.xlabel('Bioactivity',fontsize=35)
plt.ylabel('Organism',fontsize=35)
plt.show()
```
