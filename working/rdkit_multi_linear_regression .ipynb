```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from rdkit import Chem
```

```python
df = pd.read_csv('Cargill_data.csv')
df
```

```python
from rdkit.Chem import Descriptors
```

```python
mol_list= []
for element in df.canonical_smiles:
  mol = Chem.MolFromSmiles(element)
  mol_list.append(mol)
```

```python
def generate(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_MolWt = Descriptors.MolWt(mol)
        desc_NumRotatableBonds = Descriptors.NumRotatableBonds(mol)
           
        row = np.array([desc_MolLogP,
                        desc_MolWt,
                        desc_NumRotatableBonds])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MolLogP","MolWt","NumRotatableBonds"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
def generate_1(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_NumHAcceptors = Descriptors.NumHAcceptors(mol)
        desc_NumHDonors = Descriptors.NumHDonors(mol)
        desc_NumHeteroatoms = Descriptors.NumHeteroatoms(mol)
           
        row = np.array([desc_NumHAcceptors,
                        desc_NumHDonors,
                        desc_NumHeteroatoms])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["NumHAcceptors","NumHDonors","NumHeteroatoms"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
def generate_2(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolMR = Descriptors.MolMR(mol)
        desc_HeavyAtomCount = Descriptors.HeavyAtomCount(mol)
        desc_HeavyAtomMolwt = Descriptors.HeavyAtomMolWt(mol)
           
        row = np.array([desc_MolMR,
                        desc_HeavyAtomCount,
                        desc_HeavyAtomMolwt])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MolMR","HeavyAtomCount","HeavyAtomMolwt"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
def generate_3(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_NOCount = Descriptors.NOCount(mol)
        desc_NHOHCount = Descriptors.NHOHCount(mol)
        desc_NumValenceElectrons = Descriptors.NumValenceElectrons(mol)
           
        row = np.array([desc_NOCount,
                        desc_NHOHCount,
                        desc_NumValenceElectrons])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["NOCount","NHOHCount","NumValenceElectrons"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
def generate_4(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_RingCount = Descriptors.RingCount(mol)
        desc_NumAromaticCarbocyles = Descriptors.NumAromaticCarbocycles(mol)
        desc_NumAromaticHeterocycles = Descriptors.NumAromaticHeterocycles(mol)
           
        row = np.array([desc_RingCount,
                        desc_NumAromaticCarbocyles,
                        desc_NumAromaticHeterocycles])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["RingCount","NumAromaticCarbocyles","NumAromaticHeterocycles"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df1 = generate(df.canonical_smiles)
df2 = generate_1(df.canonical_smiles)
df3 = generate_2(df.canonical_smiles)
df4 = generate_3(df.canonical_smiles)
df5 = generate_4(df.canonical_smiles)
```

```python
def AromaticAtoms(m):
  aromatic_atoms = [m.GetAtomWithIdx(i).GetIsAromatic() for i in range(m.GetNumAtoms())]
  aa_count = []
  for i in aromatic_atoms:
    if i==True:
      aa_count.append(1)
  sum_aa_count = sum(aa_count)
  return sum_aa_count
```

```python
desc_AromaticProportion = [AromaticAtoms(element)/Descriptors.HeavyAtomCount(element) for element in mol_list]

```

```python
df_desc_AromaticProportion = pd.DataFrame(desc_AromaticProportion, columns=['AromaticProportion'])
```

```python
X = pd.concat([df1,df_desc_AromaticProportion,df2,df3,df4,df5], axis=1)
X
```

```python
y = df.iloc[:,3:9]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = linear_model.LinearRegression()
model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(y_train, y_pred_train))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(y_train, y_pred_train))

y_pred_test = model.predict(X_test)

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(y_test, y_pred_test))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(y_test, y_pred_test))
```
