```python
import pandas as pd
import numpy as np
```

```python
df_bos_taurus = pd.read_csv('bos_taurus.csv',sep=';') 
```

```python
df_bos_taurus.columns
```

```python
df_bos_taurus = df_bos_taurus.drop(columns=['assay_type', 'target_type', 'tax_id', 'molregno', 'molregno.1', 'chembl_id', 
                            'structure_type', 'molecule_type', 'chirality', 'natural_product', 'relation', 
                            'standard_relation', 'type', 'value', 'units'])
```

```python
seen = {}
for index, row in df_bos_taurus.iterrows():
    act_type = str(row['standard_type']) + ' /' + str(row['standard_units'])
    if act_type not in seen:
        seen[act_type] = 1
    else:
        seen[act_type] += 1

print(seen)
len(seen)
```

```python
header_list = list(seen.keys())
```

```python
for i in header_list:
        df_bos_taurus[i] = np.nan
        
for index, row in df_bos_taurus.iterrows():
    column = row['standard_type']
    df_bos_taurus.loc[index, column] = row['standard_value']
df_bos_taurus.head(n=50)
```
