```python
seen = {}
for index, row in df_cattle.iterrows():
    molregno = row['molregno']
    if molregno not in seen:
        seen[molregno] = 1
    else:
        seen[molregno] += 1

print(seen)
len(seen)
```

```python
df_cattle = df_cattle.drop(columns=['assay_type', 'target_type', 'tax_id', 'molregno', 'molregno.1', 'chembl_id', 
                                    'structure_type', 'molecule_type', 'chirality', 'natural_product', 'relation', 
                                    'standard_relation', 'type', 'value', 'units'])
df_cattle
```
