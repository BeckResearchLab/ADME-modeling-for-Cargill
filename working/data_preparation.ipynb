```python
import pandas as pd
import numpy as np
```

```python
df_cattle = pd.read_csv('cattle.csv',sep=';') 
```

```python
df_cattle
```

```python
df_cattle = df_cattle.drop(columns=['assay_type', 'target_type', 'tax_id', 'molregno', 'molregno.1', 
                                    'structure_type', 'molecule_type', 'chirality', 'natural_product', 'relation', 
                                    'standard_relation', 'type', 'value', 'units'])
df_cattle
```

```python
seen = {}
for index, row in df_cattle.iterrows():
    act_type = str(row['standard_type']) + ' /' + str(row['standard_units'])
    if act_type not in seen:
        seen[act_type] = 1
    else:
        seen[act_type] += 1

print(seen)
len(seen)
```

```python
#trying to unify the units.
header_list = ['Solubility /nM', 'T1/2 /hr', 
               'Activity /%', 'Activity /ug ml-1', 
               'IC50 /nM', 'LD50 /uM', 'ED50 /ug ml-1', 
               'F /%', 'Cmax /nM', 'Cmax /ug.mL-1', 'Tmax /hr'
               'Inhibition /%',
               'Drug metabolism /%', 'Drug metabolism /nan', 
               'Drug uptake /nan', 'Drugexcretion /%', 
               'PPB /%', 'PPB /nan', 
               'Stabilty /nan', 'Stability /%', 'Stabilty /%',
               'logKd /nan', 
               'Ratio /nan', 
               'Papp /ucm/s', "Papp /10'-6 cm/s", 'permeability /pmol/s', 'Papp /nm/s', 'permeability /%', 
               'permeability /ucm/s', 
               'AUC /uM.hr']
```

```python
for i in header_list:
        df_cattle[i] = np.nan
        
for index, row in df_cattle.iterrows():
    act_type = str(row['standard_type']) + ' /' + str(row['standard_units'])
    if act_type in header_list:
        header = act_type
        df_cattle.loc[index, header] = row['standard_value']
    else:
        df_cattle = df_cattle.drop([index])
        
df_cattle = df_cattle.sort_values(by=['chembl_id'], ascending=False).reset_index(drop=True)

df_cattle
```

```python
seen = {}
for index, row in df_cattle.iterrows():
    chembl_id = row['chembl_id']
    if chembl_id not in seen:
        seen[chembl_id] = 1
    else:
        seen[chembl_id] += 1

seen_sorted = sorted(seen.items(), key=lambda x: x[1], reverse=True) 
print(seen_sorted)
len(seen_sorted)
```

```python
df_cattle_result = pd.DataFrame(columns=df_cattle.columns)

for i, row in df_cattle.iterrows():
    if i==0:
        continue
    if row['chembl_id'] == df_cattle.loc[i-1, 'chembl_id']:
        df_cattle.iloc[i] = df_cattle.iloc[i-1].combine_first(df_cattle.iloc[i])
    else:
        df_cattle_result = df_cattle_result.append(df_cattle.iloc[i-1], ignore_index=True)

df_cattle_result = df_cattle_result.append(df_cattle.iloc[len(df_cattle)-1], ignore_index=True)


df_cattle_result

```

```python
df_cattle_result = df_cattle_result.drop(columns=['alogp', 'hba', 'hbd', 'psa', 'rtb', 'ro3_pass', 
                                                  'num_ro5_violations', 'cx_most_apka', 'cx_most_bpka', 'full_mwt', 
                                                  'cx_logp', 'cx_logd', 'molecular_species', 'mw_freebase', 
                                                  'aromatic_rings', 'heavy_atoms', 'qed_weighted', 'mw_monoisotopic', 
                                                  'full_molformula', 'hba_lipinski', 'hbd_lipinski', 
                                                  'num_lipinski_ro5_violations', 'standard_type', 'standard_value', 
                                                  'standard_units', 'standard_inchi'])
```
