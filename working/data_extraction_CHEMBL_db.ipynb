#### Defining functions for .py file

```python
def dbfile_to_dataframe (dbfile, target_organism):
    """
    Creates a SQL connection to the SQLite. This function reads SQL database files,
    executs SQL queries in python platform and extracts data for specific organism 
    and assay from database using sqlite connection by joining multiple tables, and 
    returns a pandas dataframe
    
    Parameters
    ----------
    - dbfile: path of the database file
    - target_organism: the target organism
    
    Returns
    -------
    - modified DataFrame based on the user selected tables and columns
    """
    
    con = sqlite3.connect(dbfile)
    df=pd.read_sql_query(""""  SELECT
                            a.assay_type, 
                            t.target_type, t.tax_id, t.organism,
                            act.relation, act.type, act.value, act.units, act.standard_relation, 
                            act.standard_type, act.standard_value, act.standard_units,
                            md.molregno, md.chembl_id, md.structure_type, md.molecule_type,
                            md.chirality, md.natural_product,
                            cs.standard_inchi, cs.canonical_smiles,
                            cp.*
                            FROM 
                            target_dictionary t 
                            INNER JOIN assays a on a.tid=t.tid 
                            INNER JOIN activities act on act.assay_id=a.assay_id
                            INNER JOIN compound_records cr on cr.record_id=act.record_id
                            INNER JOIN molecule_dictionary md on md.molregno=cr.molregno
                            INNER JOIN compound_structures cs on cs.molregno=md.molregno
                            INNER JOIN compound_properties cp on cp.molregno=cs.molregno
                            WHERE t.organism='target_organism'and a.assay_type='A'""", con)
    
    return df
```

```python
def merged_data(df1, df2):
    
    """This function combines two dataframes
    
    Parameters
    ----------
    - df1,df2: dataframes to be combined
    
    Returns
    -------
    - merged DataFrame
    """
    df_merged=pd.concat(df1, df2)
    
    return df_merged
```
