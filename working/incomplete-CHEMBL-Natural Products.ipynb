```python
# Import necessary libraries
import logging
from operator import itemgetter
from IPython.display import display, SVG
from chembl_webresource_client.new_client import new_client
import pandas as pd
```

```python
#retrieving all the natural products compounds filtering with no Rule-of-Five violations:

document = new_client.document
```

```python
docs = document.filter(journal="J. Nat. Prod.").only('document_chembl_id')
```

```python
compound_record = new_client.compound_record
```

```python
records = compound_record.filter(
    document_chembl_id__in=[doc['document_chembl_id'] for doc in docs]).only(
    ['document_chembl_id', 'molecule_chembl_id'])
```

```python
molecule = new_client.molecule
```

```python
natural_products = molecule.filter(
    molecule_chembl_id__in=[rec['molecule_chembl_id'] for rec in records]).only(
    'molecule_structures').filter(molecule_properties__num_ro5_violations=0)
```

```python
len(natural_products)
```

```python
df_nat_prod = pd.DataFrame.from_dict(natural_products)
```

```python
#converting dictionary to a dataframe
df_nat_prod2 = pd.DataFrame({'canonical_smiles':[], 'molfile':[], 'standard_inchi':[], 'standard_inchi_key':[]})
for i in range(len(natural_products)):    
    df_nat_prod2 = df_nat_prod2.append({'canonical_smiles': natural_products[i].get('molecule_structures').get('canonical_smiles'),
                  'molfile': natural_products[i].get('molecule_structures').get('molfile'),
                  'standard_inchi': natural_products[i].get('molecule_structures').get('standard_inchi'),
                  'standard_inchi_key': natural_products[i].get('molecule_structures').get('standard_inchi_key')},
                   ignore_index=True)
```

```python
df_nat_prod2.to_csv('natural_products_data_raw2.csv', index=False)
```

```python
# Version 2, importing csv dataframe from CHEMBL with target type:Bos taurus
df_bos=pd.read_csv('Bos_taurus_assayA.csv')
```

```python
df_bos.head()
```

```python
#getting bioactivity records for all chembl ids in  assay(A)
bioactivity=pd.DataFrame()
for chembl_id in df_bos['ChEMBL ID']:
    #print (chembl_id)
    records = new_client.activity.filter(assay_chembl_id=chembl_id)
    #print(records)
    df_records=pd.DataFrame.from_dict(records)
    bioactivity=bioactivity.append(df_records)
```

```python
bioactivity.reset_index(drop=True, inplace=True)
bioactivity
```

```python
bioactivity.to_csv('raw_data_bioactivity.csv', index=False)
```

```python
#dropping the compounds which standard values are not available 
bioact2 = bioactivity[bioactivity.standard_value.notna()]
```

```python
bioact2.reset_index(drop=True, inplace=True)
bioact2
```

```python
# retrieving bioactivity data reported as IC50 values
bioact_IC50=pd.DataFrame()
for chembl_id in df_bos['ChEMBL ID']:
    #print (chembl_id)
    records2 = new_client.activity.filter(assay_chembl_id=chembl_id).filter(standard_type='IC50')
    #print(records2)
    df_records2=pd.DataFrame.from_dict(records2)
    bioact_IC50=bioact_IC50.append(df_records2)
```

```python
bioact_IC50.reset_index(drop=True, inplace=True)
bioact_IC50
```
