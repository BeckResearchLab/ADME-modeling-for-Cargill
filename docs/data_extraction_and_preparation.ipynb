```python
bioact = pd.DataFrame()
for chembl_id in assay_A['ChEMBL ID']:
    records = new_client.activity.filter(assay_chembl_id=chembl_id)
    records_df = pd.DataFrame(list(records.all()))
    bioact = bioact.append(records_df)   
```

```python
seen = {}
for index, row in bioact.iterrows():
    activity_id = row['activity_id']
    if activity_id not in seen:
        seen[activity_id] = 1
    else:
        seen[activity_id] += 1
        print(activity_id, seen[activity_id])
```

```python
mol_act = pd.DataFrame()
j = 0
for i in range(0, len(bioact), 100):
    for ml_id in bioact[i:i+100]['molecule_chembl_id']:
        mol = molecule.get(ml_id)
        mol_df = pd.DataFrame(list(mol.items())).transpose()
        mol_df = mol_df.rename(columns=mol_df.iloc[0]).drop(mol_df.index[0])
        if mol_df.iloc[0]['biotherapeutic']!= None:
            mol_act = mol_act.append(mol_df)
            j=j+1
    print(j)
```
